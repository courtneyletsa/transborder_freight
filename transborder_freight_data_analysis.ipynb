```python
trip_counts = df_2020_2024.groupby('Mode of Transportation').size().reset_index(name='Trip Count')
plt.figure(figsize=(16, 10))
sns.barplot(data=trip_counts, x='Mode of Transportation', y='Trip Count')
plt.title('Number of Trips per Transportation Mode')
plt.xlabel('Transportation Mode')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
