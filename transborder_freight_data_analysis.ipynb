# EDA - Exploratory Data Analysis

In this section, I answer the following business questions to further explore the dataset. The questions are;

*Freight Movement Patterns*

What are the most common modes of transportation used for exports from Alaska, and how do they compare in terms of shipping weight and value?
 â†’ Helps you understand transportation mode preferences and efficiency.


Which countries (or Canadian provinces) receive the highest volume and value of exports from Alaska?
 â†’ Useful for identifying strong trade relationships and potential market expansion.


Is there any correlation between the type of container (Containerized vs Non-Containerized) and shipping costs or weight?
 â†’ Can guide container usage optimization.


_Operational Inefficiencies_

Which ports/districts handle the most freight, and do they also incur the highest freight charges?
 â†’ This can reveal bottlenecks or inefficiencies at specific ports.


Are there transportation modes or routes that consistently report zero or very low shipping weight but incur high freight charges?
 â†’ Highlights potential underutilization or inefficiencies in resource usage.


_Environmental Impact_

Which transportation modes have the highest freight charges per kilogram of weight shipped?
 â†’ Could be a proxy for energy intensity or environmental cost.


How does the distribution of shipping weights across different transportation modes inform us about their likely environmental impact?
 â†’ Lighter shipments via air, for instance, might have higher emissions per kg than heavier sea freight.


```python
# Step 1: Group by Trade Type and Mode of Transportation
summary = df_2020_2024.groupby(['Trade Type', 'Mode of Transportation']).agg({
    'Shipping Weight (kg)': 'sum',
    'Value (USD)': 'sum',
    'Mode of Transportation': 'count'  # Frequency count
}).rename(columns={'Mode of Transportation': 'Count'}).reset_index()

# Step 2: Sort for better readability (optional)
summary = summary.sort_values(by='Count', ascending=False)

# Display the summary table
summary
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(data=summary, x='Mode of Transportation', y='Value (USD)', hue='Trade Type')
plt.title('Value (USD) by Mode of Transportation (Export vs Import)')
plt.ylabel('Value (USD)')
plt.xlabel('Mode of Transportation')
plt.xticks(rotation=45)
plt.legend(title='Trade Type')
plt.tight_layout()
plt.show()
```
