The objective of this project is to analyze the transportation data provided by the Bureau of Transportation Statistics (BTS) to uncover insights related to the efficiency, safety and environmental impacts of freight transportation across various modes (road, rail, air, and water). ### Freight Movement Patterns

This area explores the common mode of transportation for imports and exports, trade relationship between these countries and container usage optimization

```python
# Group by Trade Type and Mode of Transportation
import_export_summary = df_2020_2024.groupby(['Trade Type', 'Mode of Transportation']).agg({
    'Shipping Weight (kg)': 'sum',
    'Value (USD)': 'sum',
    'Mode of Transportation': 'count'
}).rename(columns={'Mode of Transportation': 'Count'}).reset_index()

# Sort for better readability
import_export_summary = import_export_summary.sort_values(by='Count', ascending=False)
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(data=import_export_summary, x='Mode of Transportation', y='Value (USD)', hue='Trade Type')
plt.title('Value (USD) by Mode of Transportation (Export vs Import)')
plt.ylabel('Value (USD)')
plt.xlabel('Mode of Transportation')
plt.xticks(rotation=45)
plt.legend(title='Trade Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(data=import_export_summary, x='Mode of Transportation', y='Shipping Weight (kg)', hue='Trade Type')
plt.title('Shipping Weight (kg) by Mode of Transportation (Export vs Import)')
plt.ylabel('Shipping Weight (kg)')
plt.xlabel('Mode of Transportation')
plt.xticks(rotation=45)
plt.legend(title='Trade Type')
plt.tight_layout()
plt.show()
```

### Environmental impact of transborder freight movemnet

```python
#highest freight charges per kilogram of weight shipped per mode of transportation
```

```python
data_with_no_zero_shipping_weight = df_2020_2024[df_2020_2024['Shipping Weight (kg)'] > 0]
data_with_no_zero_shipping_weight['Value per kg'] = data_with_no_zero_shipping_weight['Value (USD)'] / data_with_no_zero_shipping_weight['Shipping Weight (kg)']
value_per_kg_by_mode = data_with_no_zero_shipping_weight.groupby('Mode of Transportation').agg({
    'Value per kg': 'mean',
    'Shipping Weight (kg)': 'sum',
    'Value (USD)': 'sum'
}).reset_index()
value_per_kg_by_mode = value_per_kg_by_mode.sort_values(by='Value per kg', ascending=False)

value_per_kg_by_mode
```

```python
plt.figure(figsize=(12, 10))
bars = plt.bar(value_per_kg_by_mode['Mode of Transportation'], value_per_kg_by_mode['Value per kg'], color='blue')

# Add labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}', ha='center', fontsize=9)

plt.title('Average Freight Charge per Kilogram by Transportation Mode')
plt.ylabel('Value per kg (USD)')
plt.xlabel('Mode of Transportation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

These tree modes of transportation Other, Mail and Air stand out for their extremely high cost per kg. These modes are likely used for low-volume, high-priority, or high-value goods (e.g. express mail, medical supplies, tech parts). Most often, high cost correlates with speed and convenience but also with higher energy use and emissions per kg.

Other ransportation modes such as Rail, Vessel, and Pipeline are the most environmentally friendly options slower but vastly more efficient in energy use and emissions.

Optimizing logistics for lower-cost modes not only saves money but also reduces environmental impact especially important for sustainable supply chains. As a logistics strategy, use air/mail/other only when necessary, for example if the goods are perishable or high-value urgent goods. Then for bulk, heavy cargo, it will be preferable to use more  environmentally friendly options.


```python
# distribution of shipping weights across different transportation modes
```

```python
weight_by_mode = df_2020_2024.groupby('Mode of Transportation')['Shipping Weight (kg)'].sum().reset_index()
weight_by_mode_sorted = weight_by_mode.sort_values(by='Shipping Weight (kg)', ascending=False)

plt.figure(figsize=(12,10))
bars = plt.bar(weight_by_mode_sorted['Mode of Transportation'], weight_by_mode_sorted['Shipping Weight (kg)'], color='orange')

# Add labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1e7, f'{yval/1e9:.2f}B', ha='center')
    
plt.title('Total Shipping Weight by Mode of Transportation')
plt.xlabel('Mode of Transportation')
plt.ylabel('Shipping Weight (kg)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the chart we can see that, Rail and Vessel Dominate in Shipping Weight and these modes of transporting goods are typically fuel-efficient per ton-kilometer meaning they can move a lot of weight with relatively low emissions. These modes are better for the environment when used to move heavy goods over long distances.

Trucks which is second only to Pipeline and Vessel on the other hand is efficient for short distance but but less sustainable for long-distance or heavy freight compared to rail/vessel. While flexible and fast, trucks have higher emissions per unit of cargo than rail or vessel. Additionally, Overuse of trucks for long-haul freight can increase carbon footprint which affects the environment in a bad way.
