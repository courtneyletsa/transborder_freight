```python
#replace values of some of the columns

df_2020_2024['DISAGMOT'] = df_2020_2024['DISAGMOT'].replace({
    1:'Vessel',
    3:'Air',
    4:'Mail (U.S. Postal Service)',
    5:'Truck',
    6:'Rail',
    7:'Pipeline',
    8:'Other',
    9:'Foreign Trade Zones (FTZs)'
}
)
```

```python
df_2020_2024['CONTCODE'] = df_2020_2024['CONTCODE'].replace({
    'X':'Containerized',
    '0':'Non-Containerized'
})
```

```python
df_2020_2024['TRDTYPE'] = df_2020_2024['TRDTYPE'].replace({
    1:'Export',
    2:'Import'
})
```

```python
df_2020_2024['COUNTRY'] = df_2020_2024['COUNTRY'].replace({
    1220:'Canada',
    2010:'Mexico'
})
```

```python
df_2020_2024['DF'] = df_2020_2024['DF'].replace({
    1.0:'domestically produced merchandise',
    2.0:'foreign produced merchandise'
})
```

```python
df_2020_2024['MONTH'] = df_2020_2024['MONTH'].replace({
    1.0:'January',
    2.0:'February',
    3.0:'March',
    4.0:'April',
    5.0:'May',
    6.0:'June',
    7.0:'July',
    8.0:'August',
    9.0:'September',
    10.0:'October',
    11.0:'November',
    12.0:'December'
})
```

```python
trip_counts = df_2020_2024.groupby('DISAGMOT').size().reset_index(name='Trip Count')
plt.figure(figsize=(16, 10))
sns.barplot(data=trip_counts, x='DISAGMOT', y='Trip Count')
plt.title('Number of Trips per Transportation Mode')
plt.xlabel('Transportation Mode')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
