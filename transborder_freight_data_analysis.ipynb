# TRANSBODER FREIGHT DATA ANALYSIS

The objective of this project is to analyze the transportation data provided by the Bureau of Transportation Statistics (BTS) to uncover insights related to the efficiency, safety and environmental impacts of freight transportation across various modes (road, rail, air, and water). Using this data, the goal is to identify inefficiencies, recognize patterns and propose actionable solutions to improve the overall performance and sustainability of transportation systems

### Freight Movement Patterns

This area explores the common mode of transportation for imports and exports, trade relationship between these countries and container usage optimization

```python
# Group by Trade Type and Mode of Transportation
import_export_summary = df_2020_2024.groupby(['Trade Type', 'Mode of Transportation']).agg({
    'Shipping Weight (kg)': 'sum',
    'Value (USD)': 'sum',
    'Mode of Transportation': 'count'
}).rename(columns={'Mode of Transportation': 'Count'}).reset_index()

# Sort for better readability
import_export_summary = import_export_summary.sort_values(by='Count', ascending=False)
```

```python
plt.figure(figsize=(12, 10))
ax = sns.barplot(data=import_export_summary, x='Mode of Transportation', y='Value (USD)', hue='Trade Type')

# Add data labels
for bar in ax.patches:
    height = bar.get_height()
    if height > 0:
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            height,
            f'{height:,.0f}', 
            ha='center',
            va='bottom',
            fontsize=9,
            rotation=90
        )

plt.title('Value (USD) by Mode of Transportation (Export vs Import)')
plt.ylabel('Value (USD)')
plt.xlabel('Mode of Transportation')
plt.xticks(rotation=45)
plt.legend(title='Trade Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
ax = sns.barplot(data=import_export_summary, x='Mode of Transportation', y='Shipping Weight (kg)', hue='Trade Type')

# Add data labels 
for bar in ax.patches:
    height = bar.get_height()
    if height > 0:
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            height,
            f'{height:,.0f}',  
            ha='center',
            va='bottom',
            fontsize=9,
            rotation=90
        )

plt.title('Shipping Weight (kg) by Mode of Transportation (Export vs Import)')
plt.ylabel('Shipping Weight (kg)')
plt.xlabel('Mode of Transportation')
plt.xticks(rotation=45)
plt.legend(title='Trade Type')
plt.tight_layout()
plt.show()
```

- Truck is the dominant mode for both exports and imports because they are the primary choice for cross-border trade (especially with Canada and Mexico), likely due to speed and flexibility in North America. 

- It is likely Air is less common for import because it is efficient for lightweight, high-value exports (e.g., electronics, pharmaceuticals). 

- Rail sees substantial use, especially for imports becuase heavy or bulk commodities (e.g., oil, minerals) might come in more frequently via rail.

- Transportation modes such as Trucks, Rails, Vessels and Pipelines dorminate imports because these vehicles are used in carrying massive volumes and suitable for long-distance trade. Air transport is used for lightweight, high-value items and it is extremely expensive per kilogram, so it’s not used for heavy imports.

```python
us_exports = df_2020_2024[
    (df_2020_2024['Trade Type'] == 'Export') &
    (df_2020_2024['Country'].isin(['Canada', 'Mexico']))
]

export_count_value = us_exports.groupby('Country').agg({
    'Country': 'count' ,
    'Value (USD)': 'sum'
}).rename(columns={'count': 'Export Count'})

export_count_value.columns = [
    'Export Count',
    'Total Value (USD)'
]

export_count_value_table = export_count_value.reset_index()
export_count_value_table
```

The export count of Canada is higher than Mexico and this could be as a result of the following;

- Stronger Trade Relationship with Canada (by frequency)
A higher number of exports could indicate more frequent transactions between Canada and the U.S. This may reflect closer economic integration, especially in intermediate goods (parts/materials used in U.S. manufacturing).

- Geographical and Infrastructural Advantages
Canada shares the longest land border with the U.S., and their transport infrastructure (rails, roads, pipelines) is deeply interconnected. This facilitates efficient, high-frequency exports, particularly in industries like automotive, oil, timber, and machinery.

- Trade Agreements at Play
Both Canada and Mexico are part of the USMCA (formerly NAFTA), but Canada may have more industries that align with U.S. import needs. Regulatory alignment between the U.S. and Canada (e.g. standards, inspections) often reduces friction in exporting.

### Environmental impact of transborder freight movemnet

```python
#highest freight charges per kilogram of weight shipped per mode of transportation
```

```python
data_with_no_zero_shipping_weight = df_2020_2024[df_2020_2024['Shipping Weight (kg)'] > 0]
data_with_no_zero_shipping_weight['Value per kg'] = data_with_no_zero_shipping_weight['Value (USD)'] / data_with_no_zero_shipping_weight['Shipping Weight (kg)']
value_per_kg_by_mode = data_with_no_zero_shipping_weight.groupby('Mode of Transportation').agg({
    'Value per kg': 'mean',
    'Shipping Weight (kg)': 'sum',
    'Value (USD)': 'sum'
}).reset_index()
value_per_kg_by_mode = value_per_kg_by_mode.sort_values(by='Value per kg', ascending=False)

value_per_kg_by_mode
```

```python
plt.figure(figsize=(12, 10))
bars = plt.bar(value_per_kg_by_mode['Mode of Transportation'], value_per_kg_by_mode['Value per kg'], color='blue')

# Add labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}', ha='center', fontsize=9)

plt.title('Average Freight Charge per Kilogram by Transportation Mode')
plt.ylabel('Value per kg (USD)')
plt.xlabel('Mode of Transportation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

These three modes of transportation Other, Mail and Air stand out for their extremely high cost per kg. These modes are likely used for low-volume, high-priority, or high-value goods (e.g. express mail, medical supplies, tech parts). Most often, high cost correlates with speed and convenience but also with higher energy use and emissions per kg.

Other transportation modes such as Rail, Vessel, and Pipeline are the most environmentally friendly options slower but vastly more efficient in energy use and emissions.

Optimizing logistics for lower-cost modes not only saves money but also reduces environmental impact especially important for sustainable supply chains. As a logistics strategy, use air/mail/other only when necessary, for example if the goods are perishable or high-value urgent goods. Then for bulk, heavy cargo, it will be preferable to use more  environmentally friendly options.


```python
# distribution of shipping weights across different transportation modes
```

```python
weight_by_mode = df_2020_2024.groupby('Mode of Transportation')['Shipping Weight (kg)'].sum().reset_index()
weight_by_mode_sorted = weight_by_mode.sort_values(by='Shipping Weight (kg)', ascending=False)

plt.figure(figsize=(12,10))
bars = plt.bar(weight_by_mode_sorted['Mode of Transportation'], weight_by_mode_sorted['Shipping Weight (kg)'], color='orange')

# Add labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1e7, f'{yval/1e9:.2f}B', ha='center')
    
plt.title('Total Shipping Weight by Mode of Transportation')
plt.xlabel('Mode of Transportation')
plt.ylabel('Shipping Weight (kg)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the chart we can see that, Rail and Vessel Dominate in Shipping Weight and these modes of transporting goods are typically fuel-efficient per ton-kilometer meaning they can move a lot of weight with relatively low emissions. These modes are better for the environment when used to move heavy goods over long distances.

Trucks which is second only to Pipeline and Vessel on the other hand is efficient for short distance but but less sustainable for long-distance or heavy freight compared to rail/vessel. While flexible and fast, trucks have higher emissions per unit of cargo than rail or vessel. Additionally, Overuse of trucks for long-haul freight can increase carbon footprint which affects the environment in a bad way.

### Operational inefficiencies of  transborder freight 

### Are certain transportation modes more cost-efficient than others in terms of freight charges per kg of goods shipped?

```python
data_with_no_zero_shipping_weight = df_2020_2024[df_2020_2024['Shipping Weight (kg)'] > 0]
data_with_no_zero_shipping_weight['Cost_per_kg'] = data_with_no_zero_shipping_weight['Freight Charges (USD)'] / data_with_no_zero_shipping_weight['Shipping Weight (kg)']
data_with_no_zero_shipping_weight = data_with_no_zero_shipping_weight.dropna(subset=['Mode of Transportation', 'Cost_per_kg'])

plt.figure(figsize=(12, 10))
sns.boxplot(data=data_with_no_zero_shipping_weight, x='Mode of Transportation', y='Cost_per_kg')
plt.title('Cost Efficiency by Mode of Transportation (Freight Charges per kg)')
plt.ylabel('Freight Charges per kg (USD)')
plt.xlabel('Mode of Transportation')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

The distribution of freight charges per kg across different transportation modes is a way to determine which transportation modes more cost-efficient. Modes with lower median values and fewer extreme outliers (like Vessel and Pipeline) appear more cost-efficient, while modes like Air and Other show higher median costs and greater variability, indicating they are less cost-efficient for shipping goods by weight.
